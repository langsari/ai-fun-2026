```python
from google import genai

client = genai.Client(api_key="YOUR_API_KEY")

data = {
    "sleep_hours": 8,
    "energy_level": 9,
    "muscle_soreness": 2,
    "stress_level": 2,
    "consecutive_workout_days": 1,
    "weight": 55,
    "height": 165,
    "gender": "female",
    "body_fat": 24
}

bmi = data["weight"] / ((data["height"] / 100) ** 2)

prompt = f"""
You are an AI Fitness Coach.

User Data:
Sleep: {data['sleep_hours']} hours
Energy: {data['energy_level']}/10
Soreness: {data['muscle_soreness']}/10
Stress: {data['stress_level']}/10
Workout days: {data['consecutive_workout_days']}
BMI: {bmi:.2f}

Decide:
1. Rest / Light / Intense
2. Explain why
3. Suggest workout
"""

try:
    response = client.models.generate_content(
        model="gemini-2.0-flash",
        contents=prompt
    )
    result = response.text

except Exception as e:
    print("⚠️ Using fallback system...")

    # --------- SCORING SYSTEM ----------
    score = 0

    # Sleep
    if data["sleep_hours"] >= 7:
        score += 2
    elif data["sleep_hours"] >= 5:
        score += 1
    else:
        score -= 2

    # Energy
    if data["energy_level"] >= 7:
        score += 2
    elif data["energy_level"] >= 5:
        score += 1
    else:
        score -= 1

    # Muscle soreness
    if data["muscle_soreness"] >= 7:
        score -= 2
    elif data["muscle_soreness"] >= 4:
        score -= 1
    else:
        score += 1

    # Stress
    if data["stress_level"] >= 7:
        score -= 2
    elif data["stress_level"] >= 4:
        score -= 1
    else:
        score += 1

    # Workout days
    if data["consecutive_workout_days"] >= 4:
        score -= 2
    elif data["consecutive_workout_days"] >= 2:
        score -= 1
    else:
        score += 1

    # --------- DECISION ----------
    if score <= 0:
        result = f"""
=== AI RESULT (Fallback) ===

Decision: Rest

Reason:
Body is fatigued or not fully recovered.

Recommendation:
- Take full rest
- Sleep at least 7-8 hours
- Light stretching only

Score: {score}
"""
    elif score <= 3:
        result = f"""
=== AI RESULT (Fallback) ===

Decision: Light Workout

Reason:
Moderate condition but not optimal.

Recommendation:
- Light cardio (20-30 mins)
- Stretching
- Avoid heavy lifting

Score: {score}
"""
    else:
        result = f"""
=== AI RESULT (Fallback) ===

Decision: Intense Workout

Reason:
Body is in good condition and ready.

Recommendation:
- Strength training
- Cardio 30 mins
- High intensity workout

Score: {score}
"""

# -------------------------
# OUTPUT
# -------------------------
print(result)
```
